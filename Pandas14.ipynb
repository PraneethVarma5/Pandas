```python
import pandas as pd
```

```python
df=pd.read_csv('Pandas_Time_Series2.csv')
df#here we dont have dates in our dataset
```

```python
rng=pd.date_range(start="6/1/2017",end="6/30/2017",freq='B')
rng#here it creates a date range from 6-1-2017 to 6-30-2017
#And in frequency B is mentioned which means business days which will exclude weekends
```

```python
df.set_index(rng,inplace=True)
df
```

```python
df.loc['2017-06-01']
```

```python
df.asfreq('D',method='pad')
#D - daily
#this includes business and weekends also
#pad is used to carry forward the values to another row
#observe 2nd 3rd and 4th have same values 
```

```python
df.asfreq('W',method='pad')
#W-weekly prices
```

```python
df.asfreq('H',method='pad')
#H-Hourly prices
```

```python
rng=pd.date_range(start='1-1-2017',periods=72,freq='B')
rng
#creates a date range from jan 1st 2017 upto 72 days without weekends
```
