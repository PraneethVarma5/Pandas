```python
''' Cross tab is also known as cross tabulation or contingency table is a type
of table in a matrix form that displays the frequance distribution of the variables.
In simple terms it gives a main brief info in another table which is useful in data analytics
'''
```

```python
import pandas as pd
```

```python
df=pd.read_excel('survey_data_for_crosstabs.xls')
df
```

```python
df1=pd.crosstab(df.Nationality,df.Handedness)
df1#it creates a crosstab with nationality as row and handedness as columns
#so here we can see the count of left handed and right handed from the original data
#in crosstab, these are called frequancy
```

```python
df1=pd.crosstab(df.Sex,df.Handedness)
df1
```

```python
df1=pd.crosstab([df.Sex,df.Nationality],df.Handedness,margins=True)
df1
```

```python
df1=pd.crosstab(df.Sex,df.Handedness,normalize="index")
df1#gives the percentage
```

```python
import numpy as np
df1=pd.crosstab(df.Sex,df.Handedness,values=df.Age,aggfunc=np.average)
df1#it is used to get the average age based on their handedness and sex
#so in female avg age is 44.5 who are left handed and 31.0 is avg age who are right handed
```
