```python
import pandas as pd
```

```python
df=pd.read_csv('sample_stock_data.csv')
df
```

```python
df=pd.read_csv('sample_stock_data.csv',skiprows=1) #it means it skips the 1st row 
df
```

```python
df=pd.read_csv('sample_stock_data.csv',header=1)#it means the heading starts from 1st line and in excel 1st line means 2nd one due to indexing
df
```

```python
df=pd.read_csv('sample_stock_data.csv',header=None) #if you dont have headings then it will mention index as headers starting from 0.
df
```

```python
df=pd.read_csv('sample_stock_data.csv',header=None,names=['heading in order'])# if you dont have headings then it will mention index as headers starting from 0. And names is used to assign headings as you like
df
```

```python
df=pd.read_csv('sample_stock_data.csv',nrows=3)# it is used to take first 3 rows and store it in df
df
```

```python
df=pd.read_csv('sample_stock_data.csv',na_values=['not available','n.a.']) #in this na_values assigns all the mentioned statements with NaN(not a number) values
df
```

```python
df = pd.read_csv("sample_stock_data.csv",  na_values={'eps': ['not available'],'revenue': [-1],'people': ['not available','n.a.']}) #in this as we can see the revenue is -1 as it cant be negative so if we mention directly like previous one all the -1 values will be change but we have to change only revenue, so we use dictionary for this 
df
```

```python
df.to_csv('new.csv',index=False) # it creates a new csv file with "new" as name with the content in df and index=False means the index values 0-4 will be reomved in csv file 
```
