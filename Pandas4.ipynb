```python
df=pd.read_csv("missing_weather_data.csv",parse_dates=["day"])
df
```

```python
type(df.day[0])
```

```python
df.set_index('day',inplace=True)
df
```

```python
df1=df.fillna(0) #it fills all the NaN values with 0
df1
```

```python
df1=df.fillna({'temperature':0,'windspeed':0,'event':'no event'}) #here it will fill temp and windspeed with 0 and event with no event using dict
df1
```

```python
df1=df.fillna(method="ffill") #it will replace the NaN values with previous values in same row
df1
```

```python
df1=df.fillna(method="bfill") #it will replace the NaN values with next values in same row
df1
```

```python
df1=df.fillna(method="bfill",axis="columns") #it will replace the NaN values with previous values in same column
df1
```

```python
df1=df.fillna(method="ffill",limit=1)#here limit=1 means if there are 2 NaN values one below the other it fills only one value
df1
```

```python
df1=df.interpolate()#interpolate is method which fills the missing series values middle values of the previous and next value.By default it is linear df1
df1
```

```python
df1=df.dropna()#it deletes all the rows with NaN
df1
```

```python
df1=df.dropna(how="all")#it deletes the rows which has all NaN values in a row
df1
```

```python
df1=df.dropna(thresh=2)#it means it doesn't delete row which has 2 or more not NaN value. deletes remaining
df1
```
