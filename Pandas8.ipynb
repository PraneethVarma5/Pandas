```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","san fransico"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city")
df3#only common city in both dataframes will be merged it is same as inner join
```

```python
df3=pd.merge(df1,df2,on="city",how="outer")
df3#it joins or combines all the values in the dataframe
```

```python
df3=pd.merge(df1,df2,on="city",how="left")
df3#when we mention left it takes all the values in left table i.e df1 and common values in df2
```

```python
df3=pd.merge(df1,df2,on="city",how="right")
df3#when we mention right it takes all the values in right table i.e df2 and common values in df1
```

```python
df3=pd.merge(df1,df2,on="city",how="outer",indicator="True")
df3#we get know that from which df the values came by using indicator
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3= pd.merge(df1,df2,on="city",how="outer")
df3
```

```python
df3= pd.merge(df1,df2,on="city",how="outer", suffixes=('_first','_second'))
df3
```
