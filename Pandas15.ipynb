```python
from pandas.tseries.holiday import USFederalHolidayCalendar
from pandas.tseries.offsets import CustomBusinessDay
#The USFederalHolidayCalendar and CustomBusinessDay tools in pandas are quite useful when working with date ranges that need to account for specific holidays or custom business days
#IndianHolidayCalendar from the pandas_market_calendars library to handle Indian holidays.
```

```python
usb=CustomBusinessDay(calendar=USFederalHolidayCalendar())
usb
```

```python
rng=pd.date_range(start='7-1-2017',end='8-2-2017',freq=usb)
```

```python
df.set_index(rng,inplace=True)
df
```
