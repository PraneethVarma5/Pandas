```python
data={
    'day':['1-7-24','2-7-24','3-7-24'],
    'temp':[32,25,38],
    'event':['Rainy','Sunny','Snow']
}
df=pd.DataFrame(data)
df #way to create a dataframe using python dictionary
```

```python
data=[
    ('1-7-24',32,6,'Rainy'),
    ('2-7-24',25,8,'Snow'),
    ('3-7-24',38,5,'Sunny')
]
df=pd.DataFrame(data,columns=['day','temp','windspeed','event'])
df
# way to create DF using list of tuples
```

```python
data=[
    {'day':'1-7-24','temp':32,'event':'Rainy'},
    {'day':'2-7-24','temp':25,'event':'Snow'},
    {'day':'3-7-24','temp':38,'event':'Sunny'},
]
df=pd.DataFrame(data)
df # way to create DF using list of dictionaries 
```
