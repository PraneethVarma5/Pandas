```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("handling_missing_data_using_replace.csv")
df
```

```python
df=df.replace({'temperature':'[A-Za-z]',
               'windspeed':'[A-Za-z]'},'',regex=True)
df#it removes all the units in dataset using regex method(Regular Expression)
```

```python
df1=df.replace('-99999',np.NaN)#it means it replaces one value with the other value
df1
```

```python
df1=df.replace({'event':'0','temperature':'-99999','windspeed':'-99999'},np.NaN)#Here instead of taking all rows and cols at a time, it checks each row with desired input and changes it as NaN
df1
```

```python
df1=df.replace({'-99999':np.NaN,'0':'sunny'})#it is a mapping method which changes the values 
df1
```

```python
df = pd.DataFrame({
    'score': ['exceptional','average', 'good', 'poor', 'average', 'exceptional'],
    'student': ['rob', 'maya', 'parthiv', 'tom', 'julian', 'erica']
})
df
```
